### Load data

```python
def load_data(path, type):
    path = os.path.join(dir, path)
    with open(path) as f:
        lines = f.readlines()
        

    if type == "mm":
        data = lines[29:]
        data = [line.split() for line in data]
        df = pd.DataFrame(data, columns=["time", "vdw_protein", "elec_protein", "vdw_ligand", "elec_ligand", "vdw_complex", "elec_complex", "total_energy"]).astype(float)
    elif type == "polar":
        data = lines[25:]
        data = [line.split() for line in data]
        df = pd.DataFrame(data, columns =["time", "protein", "ligand", "complex"]).astype(float)
    elif type == "nonpolar":
        data = lines[31:]
        data = [line.split() for line in data]
        data = [line[:4] for line in data]
        df = pd.DataFrame(data, columns = ["time", "protein", "ligand", "complex"]).astype(float)
    else:
        raise ValueError("type must be either mm, polar, or nonpolar")
    
    return df
```

### Calculate G

calculate G value for each parameter in each row dataset

```python
df_calc["G_protein"] = df_calc["mm_mean_protein"] + df_calc["polar_protein"] + df_calc["npolar_protein"]
df_calc["G_ligand"] = df_calc["mm_mean_ligand"] + df_calc["polar_ligand"] + df_calc["npolar_ligand"]
df_calc["G_complex"] = df_calc["mm_mean_complex"] + df_calc["polar_complex"] + df_calc["npolar_complex"]
```

calculation result in dataset

```python
df_calc.head()
```

### Calculate Gbinding

calculate Gbindig for each row

```python
df_calc["G_binding"] = df_calc["G_complex"] - df_calc["G_protein"] - df_calc["G_ligand"]
```

```python
G_binding_mean = np.mean(df_calc["G_binding"])
```

```python
G_binding_std = np.std(df_calc["G_binding"])
```
